```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('populations.txt', dtype ={'names': ('year', 'hare', 'lynx', 'carrot'), 
                                  'formats': ('i4', 'f4', 'f4', 'f4')})
```

$1.$

```python
plt.plot(data['year'], data['hare'])
plt.plot(data['year'], data['lynx'])
plt.plot(data['year'], data['carrot'])
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(('hare', 'lynx', 'carrot'))
plt.show()
```

$2.$

```python
a = data['lynx'] / (data['lynx'] + data['hare']) * 100
plt.bar (data['year'], a)
plt.bar (data['year'], 100-a,  bottom=a)
plt.xlabel('Year')
plt.ylabel('Percent')
plt.legend(('hare', 'lynx'))
plt.show()
```

$3.$

```python
print('Mean and std for lynx: {}, {}'.format(np.mean(data['lynx']), np.std(data['lynx'])))
print('Mean and std for hare: {}, {}'.format(np.mean(data['hare']), np.std(data['hare'])))
print('Mean and std for carrot: {}, {}'.format(np.mean(data['carrot']), np.std(data['carrot'])))

plt.hist(data['lynx'], 10)
plt.axvline(x=np.mean(data['lynx']), color = 'red')
plt.title('lynx')
plt.show()

plt.hist(data['hare'], 10)
plt.axvline(x=np.mean(data['hare']), color = 'red')
plt.title('hare')
plt.show()

plt.hist(data['carrot'], 10)
plt.axvline(x=np.mean(data['carrot']), color = 'red')
plt.title('carrot')
plt.show()
```

$4.$

```python
print('Max year for lynx: {}'.format(data['year'][data['lynx'].argmax()]))
print('Max year for hare: {}'.format(data['year'][data['hare'].argmax()]))
print('Max year for carrot: {}'.format(data['year'][data['carrot'].argmax()]))

plt.plot(data['year'], data['hare'], 'b')
plt.plot(data['year'], data['lynx'], 'r')
plt.plot(data['year'], data['carrot'], 'g')
plt.axvline(x=data['year'][data['hare'].argmax()], color='b', linestyle='--')
plt.axvline(x=data['year'][data['lynx'].argmax()], color='r', linestyle='--')
plt.axvline(x=data['year'][data['carrot'].argmax()], color='g', linestyle='--')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(('hare', 'lynx', 'carrot'))
plt.show()
```

$5.$

```python
a = np.sign(data['hare'] - data['lynx'])
b = map(lambda x: 'H' if x == 1 else 'L', a)
plt.pie([b.count('H'),b.count('L')], labels=['H', 'L'], autopct='%1.f')
plt.show()
```

$6.$

```python
print('Population > 50000 for hare: {}'.format(data['year'][np.where(data['hare'] > 50000)]))
print('Population > 50000 for lynx: {}'.format(data['year'][np.where(data['lynx'] > 50000)]))
print('Population > 50000 for carrot: {}'.format(data['year'][np.where(data['carrot'] > 50000)]))

plt.plot(data['year'], data['hare'], 'b')
plt.plot(data['year'], data['lynx'], 'r')
plt.plot(data['year'], data['carrot'], 'g')

plt.vlines(data['year'][np.where(data['hare'] > 50000)], 0, 90000, color='b', linestyle='--', alpha=0.5)
plt.vlines(data['year'][np.where(data['lynx'] > 50000)], 0, 90000, color='r', linestyle='--',  alpha=0.5)
plt.vlines(data['year'][np.where(data['carrot'] > 50000)], 0, 90000, color='g', linestyle='--',  alpha=0.5)

plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(('hare', 'lynx', 'carrot'))
plt.axis([1900, 1920, 0, 90000])
plt.show()
```

$7.$

```python
print('Top 2 years with lowest population for hare: {}'.format(data['year'][np.argsort(data['hare'])[0:2]]))
print('Top 2 years with lowest population for lynx: {}'.format(data['year'][np.argsort(data['lynx'])[0:2]]))
print('Top 2 years with lowest population for carrot: {}'.format(data['year'][np.argsort(data['carrot'])[0:2]]))
```
